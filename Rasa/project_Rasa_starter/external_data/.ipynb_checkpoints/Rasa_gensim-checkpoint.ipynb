```python
from gensim.models import Word2Vec
import pandas as pd
import os
```

```python
combined_reviews = pd.read_csv('reviews_combined_amazon.csv',index_col=0)
```

```python
combined_reviews.head()
```

```python
combined_reviews['review_title_text'] = combined_reviews['review_title'].apply(lambda x:str(x)) + " " + combined_reviews['review_text'].apply(lambda x:str(x))
```

```python
reviews_title_text_list = combined_reviews['review_title_text'].tolist()
```

```python
reviews_title_text_list[:10]
```

```python
sentence_stream = [doc.split(' ') for doc in reviews_title_text_list]
```

```python
sentence_stream[:5]
```

```python
from gensim.models import Phrases
from gensim.models.phrases import Phraser
bigram = Phrases(sentence_stream, min_count=2, threshold=10, delimiter=b' ')
bigram_phraser = Phraser(bigram)
```

```python
tokens_list = []
for sent in sentence_stream:
    _tokens = bigram_phraser[sent]
    tokens_list.append(_tokens)
```

```python
tokens_list
```

```python
model = Word2Vec(tokens_list, size=200, window=5)
```

```python
model.wv.most_similar('phone')
```

```python
model.wv.save_word2vec_format('word2vec_shopping.txt')
```

```python
import gzip

fp = open('word2vec_shopping.txt', 'rb')
data = fp.read()

bindata = bytearray(data)

with gzip.open('word2vec_shopping.txt.gz', 'wb') as f:
    f.write(bindata)
```
